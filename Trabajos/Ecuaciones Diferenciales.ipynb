# Ejercicio 1

$(1+e^{x})yy^{'} = e^{x}$

Halla la solución que pasa por (0,1)

```python
f = lambda x: np.sqrt(2*np.log(1+np.exp(x))-0.3862943)
x2 = np.linspace(0,3,100)
l = lambda x,y: np.exp(x)/((1+np.exp(x))*y)
n = 50
mu = 0.002
x = np.linspace(0,3,n)
y = Ed.trapecio(l,[0,3],1,nptos = n)
plt.figure()
plt.plot(x2[::4],f(x2[::4]),'b--', label = 'Salucion analitica')
plt.plot(x,y, 'g--', label = 'aproximación')
plt.legend()
plt.grid()
```

# Ejercicio 2

$ \dfrac{dy}{dx} = \dfrac{x-y}{x}, y(1) = 1$

Compare con la solución analitica

$ y = \dfrac{x}{2} - \dfrac{C}{x} $


```python
n = 50
l = lambda x: 0.5*x-0.5/x 
x2 = np.linspace(1,3,n)
f = lambda x,y: (x - y)/x

x = np.linspace(1,2,n)
y = Ed.Runge_kutta(f,[1,2],0.5,nptos = n)
print(y)
plt.figure()
plt.plot(x2[::4],l(x2[::4]),'b--', label = 'Salucion analitica')
plt.plot(x,y, 'g--', label = 'aproximación')
plt.grid()
```

# Ejercicio 3
Defina la condición de frontera que considere pertinente
$ xy^{'} + 4y = x^{3}-x $

$y^{'} = (x^{2}-1) - \dfrac{4y}{x}$

Compare con la solucion analitica:
$y = \dfrac{x^{3}}{7} - \dfrac{x}{5} + \dfrac{C}{x^{4}}$


```python
f = lambda x,y: (x**2 - 1) - (4*y)/x
f2 = lambda x: x**3/7 - x/5 
n = 50
x = np.linspace(0,3,n)
y = Ed.Euler_izq(f,[0,3],0,nptos = n)
plt.figure()
plt.plot(x,f2(x), 'b--', label = 'aproximación')
plt.plot(x,y, 'g--', label = 'aproximación')
plt.grid()
```
