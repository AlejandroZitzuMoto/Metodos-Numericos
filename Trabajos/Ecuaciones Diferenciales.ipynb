# Ejercicio 3
Defina la condici√≥n de frontera que considere pertinente
$ xy^{'} + 4y = x^{3}-x $

$y^{'} = (x^{2}-1) - \dfrac{4y}{x}$

Compare con la solucion analitica:
$y = \dfrac{x^{3}}{7} - \dfrac{x}{5} + \dfrac{C}{x^{4}}$

Donde C $= 0$

```python
f = lambda x,y: (x**2 - 1) - (4*y)/x
f2 = lambda x: x**3/7 - x/5 
n = 50
x = np.linspace(0,3,n)
y = Ed.trapecio_disc(f,[0,3],0,nptos = n)
plt.figure()
plt.plot(x,f2(x), 'b--', label = 'Funcion real')
plt.plot(x,y, 'g--', label = 'aproximaci√≥n')
plt.legend()

plt.grid()
```

# Ejercicio 4
La condicion de frontera es $y(0) = 1$

$(x^{3} + xy^{2})dx + (x^{2}y + y^{3})dy = 0$

$\dfrac{dy}{dx} = -\dfrac{x^{3}+xy^{2}}{x^{2}y+y^{3}}$

De esta manera lo que obtendremos es la ecuacion para graficar la funcion.

Integrando la ecuacion obtenemos la funci√≥n:

$(x^{2}+y^{2})^{2} = C$ 

Donde $C = 1$

```python
f = lambda x,y: -(x**3 + x*y**2)/(x**2*y + y**3)
f2 = lambda x: np.sqrt(1 - x**2) 
n = 50
x = np.linspace(0,1,n)
y = Ed.trapecio_disc(f,[0,1],1,nptos = n)
plt.figure()
plt.plot(x,f2(x), 'b--', label = 'Funcion real')
plt.plot(x,y, 'g--', label = 'aproximaci√≥n')
plt.legend()

plt.grid()
```
