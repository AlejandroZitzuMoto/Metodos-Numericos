# Interpolacio

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generatedata(n, f, nodes="cheb"):
    '''
    Funcion para generar puntos de una función.
    ---------------------------------------------
    n: Numero de puntos
    f: Funcion
    nodes: Ni idea
    '''
    if nodes=="cheb":
        dataxs = -np.cos(np.linspace(0,np.pi,n))
    else:
        dataxs = np.linspace(-1,1,n) 
    
    datays = f(dataxs)
    return dataxs, datays
```

```python
def lagrenge(x,xv,k):
    lx = 1
    for i in range(len(xv)):
        if k != i:
            lx = lx * ((x-xv[i])/(xv[k]-xv[i]))
    return lx

def poli_cardinal(xv,yv):
    '''
    Interpolación de Legendre
    ---------------------------
    xv: conjunto de datos en x
    yv: Conjunto de datos en y
    --------------------------
    Salida: Conjunto de datos 
    '''
    n = len(xv)
    px = []
    x = np.linspace(xv[0],xv[-1],len(xv))
    for k in range(n):
        p=0
        for j in range(n-1):
            if k != j:
                p = p + yv[k]*lagrenge(x[j],xv,k)
        px.append(p)
    return x, px
```

```python
fig, axes = plt.subplots(1,3,figsize=(10,3))
# default grid appearance
y = lambda x: 1/(1+25*x**2)
x1,y1 = generatedata(15,y)
x2,y2 = poli_cardinal(x1,y1)
axes[0].plot(x1, y(x1),lw=2)
axes[0].set_xlim([-1,1])
axes[0].set_title('Funcion real')
axes[0].grid(True)

axes[1].plot(x1,y1, 'r--')
axes[1].set_xlim([-1,1])
axes[1].set_ylim([0,1])
axes[1].set_title('Puntos Aleatorios')
axes[1].grid(True)

axes[2].plot(x2, y2, 'g--',lw=2)
axes[2].set_title('Interpolacion de Lagrenge')
axes[2].grid(True)
```
