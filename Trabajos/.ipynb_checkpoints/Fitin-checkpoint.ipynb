```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def r(x,y, x0,y0):
    n = x.size
    xp = np.sum(x)/n
    yp = np.sum(y)/n
    r = (x0-xp)*(y0-yp)
    r2 = (np.sqrt((x0-xp)**2*(y0-yp)**2))
    
    return r/r2
```

# Ejercicio 1

```python
x = np.array([34, 27, 65, 20, 53, 49, 42, 31, 55, 61])
y = [4.5, 3, 7, 3.5, 8, 5, 4, 4, 5.5, 7.5]
x.size
```

```python
def fbase(n,i,x):    
    if n == 0:
        val = i
    else:
        val = x**i
    return val
```

```python
def Gnormalfit(dataxs, datays, datasigs, fb,n = 1):
    N = dataxs.size
    A = np.zeros((N, n))  # notar la dimensión
    
    for k in range(n):
        A[:, k] = fb(n,k, dataxs)/datasigs  # A_{jk} = phi_k(x_j)/sigma_j
    bs = datays/datasigs  # b_j = y_j/sigma_j
    
    matI = A.T@A  # np.dot(A.T, A)
    InvmatI = np.linalg.inv(matI) 
    matD = A.T@bs
    cs = InvmatI@matD
    
    sigS = np.diagonal(InvmatI)
    chisq = np.sum((bs - A@cs)**2)  # (bs - A@cs).T@(bs - A@cs)
    
    return cs, chisq, sigS
```

```python
N = 10
f = lambda x,c0, c1: c0 + c1*x
np.random.seed(45379)  # fijando la semilla
sigs = 0.2*np.abs(np.random.randn(N))
c0, sig, chi2 = Gnormalfit(x,y,sigs,fbase,2)
d0 , sig2, ch2 = Gnormalfit(x,y,sigs,fbase2,3)
print("El coeficiente de corelación para personas de 25 años es: %0.2f"% r(x,y,25,f(25,c0[0],c0[1])))
print("El coeficiente de corelación para personas de 70 años es: %0.2f"% r(x,y,70,f(70,c0[0],c0[1])))
print("Tiene una corelación positiva perfecta")
print()
print("El coeficiente de corelación para personas de 42 años es: %0.2f"% r(x,y,42,f(42,c0[0],c0[1])))
print("El coeficiente de corelación para personas de 31 años es: %0.2f"% r(x,y,31,f(31,c0[0],c0[1])))
print("No existe corelación lineal")
```

```python
xv = np.linspace(20,70,100)
fig, ax = plt.subplots()
ax.errorbar(x,y,xerr = None, yerr = sigs, capsize=5, ls='', marker='o')
ax.plot(xv, f(xv,c0[0],c0[1]), ls='-', lw=1, c='g')
ax.plot(xv, f2(xv,d0[0],d0[1],d0[2]), ls='-', lw=1, c='g')
ax.fill_between(xv, f(xv, c0[0],(c0[1]+sigs[2])), f(xv, c0[1],(c0[1]-sigs[1])), color='gray', alpha=0.5)
```

```python
ventas = [9, 6, 12, 7, 5, 10, 8, 4, 7, 10, 14]
demanda = np.array([200, 220, 400, 330, 210, 390, 280, 140, 280, 290, 380])
```

```python
def fbase2(n,i,x):
    if i ==0:
        val = 1
    elif i == 1:
        val = x
    else:
        val = x**2
    return val
```

```python
N = 11
f = lambda x,c0, c1: c0 + c1*x 
f2 = lambda x,c0,c1,c2: c0 + c1*x + c2*x**2
np.random.seed(45379)  # fijando la semilla
sigs = 0.2*np.abs(np.random.randn(N))
c0, sig, chi2 = Gnormalfit(demanda,ventas,sigs,fbase,2)
d0 , sig2, ch2 = Gnormalfit(demanda,ventas,sigs,fbase2,3)
```

```python
for i in range(N):
    l = demanda[i]
    print("Corelacion entre las variables. No %d, r: %0.4f"% (i,r(demanda,ventas,l,f2(l,d0[0],d0[1],d0[2]))))
print("Existe una correlaciónzexepto en la No. 9")
```
